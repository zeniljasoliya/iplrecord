```python
import pandas as pd
import matplotlib.pyplot as plt

# Calculate counts for each inning
inning_counts = df['inning'].value_counts()

# Plotting the bar chart
plt.bar(inning_counts.index, inning_counts.values)

# Adding labels and title
plt.xlabel('Inning')
plt.ylabel('Count')
plt.title('Count of Innings')

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Plotting the line chart
plt.plot(df['over'], df['total_runs'], marker='o', linestyle='-')

# Adding labels and title
plt.xlabel('Over')
plt.ylabel('Total Runs')
plt.title('Total Runs Scored over Different Overs')

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Plotting the line chart
plt.plot(df['inning'], df['extra_runs'], marker='o', linestyle='-')

# Adding labels and title
plt.xlabel('Innings')
plt.ylabel('Extra Runs')
plt.title('Extra Runs Scored Vs Innings')

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is your DataFrame containing the dataset
# Example DataFrame

# df = pd.DataFrame(data)

# Filter numeric columns
numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns

# Plot histograms for numeric columns
for column in numeric_columns:
    plt.figure(figsize=(8, 6))
    plt.hist(df[column], bins=10, edgecolor='black', color='skyblue')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.title(f'Distribution of {column}')
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Filter categorical columns
categorical_columns = df.select_dtypes(include=['object']).columns

# Plot bar charts for categorical columns
for column in categorical_columns:
    plt.figure(figsize=(8, 6))
    df[column].value_counts().plot(kind='bar', color='skyblue')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.title(f'Bar Chart of {column}')
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Assuming 'dismissal_kind' is the column for which you want to create a pie chart
column_to_plot = 'batting_team'

# Count occurrences of each category
category_counts = df[column_to_plot].value_counts()

# Plotting the pie chart
plt.figure(figsize=(8, 6))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=140)
plt.title(f'Pie Chart of {column_to_plot}')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Display the plot
plt.show()
```
